**Tokenization** (Splitting into basic units) :

Break the text into elementary units => in our case we choose the unit to be a word .

```python
import re
from typing import List

def tokenize(text: str) -> List[str]:

    # Replace apostrophes with space
    text = re.sub(r"[’'.]", " ", text)

    # Collapse multiple spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Split and remove empty tokens
    tokens = [token for token in text.split() if token]

    return tokens
```

```python
from typing import List

def filter_tokens(tokens: List[str], stopwords: List[str] = stopwords) -> List[str]:
    """Filters out French stopwords and numeric tokens from a list of tokens."""
    filtered = [ 
        token for token in tokens
        if token.lower() not in stopwords and not token.isdigit()
    ]
    return filtered
```

**Text Normalization**:


Remove punctuation: .,!? etc.

Remove accents (normalize letters): é -> e

Unify case: convert all words to lowercase

Format dates/numbers if necessary

```python
import unicodedata

def normalize_text(text: str) -> str:
    """
    Normalizes text by removing punctuation, lowercasing, removing accents.
    """
    # 1. Remove punctuation
    text = re.sub(r"[.,;:!?()\[\]\"']", "", text)
    if text.strip() == "":
        # text is empty after punctuation removal
        return None
    
    # 2. Lowercase
    text = text.lower()

    # 3. Remove accents : 
    ## Decomposes accented characters into base + accent.
    ## Removes all accents (diacritics).
    ## Joins the remaining characters back into a plain ASCII-like string.
    text = ''.join(
        c for c in unicodedata.normalize('NFD', text)
        if unicodedata.category(c) != 'Mn'
    )

    return text
```

```python
from collections import Counter
from typing import List, Tuple

def count_tokens_corpus(tokens: List[List[str]]) -> Tuple[int, int, Counter]:
    """
    Counts total and unique tokens for a corpus where each document is a list of tokens.

    Returns:
    - total_tokens: total number of tokens
    - unique_tokens: number of unique tokens
    - token_freq: frequency of each token
    """
    # Flatten the list of lists
    all_tokens = [token for doc in tokens for token in doc]

    total_tokens = len(all_tokens)
    token_freq = Counter(all_tokens)
    unique_tokens = len(token_freq)

    return total_tokens, unique_tokens, token_freq
```

```python
def preprocess_document(doc: str) -> List[str]:
    """ Preprocesses a single document: tokenization, stopword removal, normalization, lemmatization """
    tokens = tokenize(doc)
    filtered = filter_tokens(tokens)
    normalized = [normalize_text(token) for token in filtered]
    stemmed = stem_tokens(normalized)
    return stemmed

def preprocess_corpus(corpus: List[str]) -> List[List[str]]:
    """ Preprocesses a corpus of documents: tokenization, stopword removal, normalization, lemmatization """
    preprocessed_corpus = []
    for doc in corpus:
        stemmed = preprocess_document(doc)
        preprocessed_corpus.append(stemmed)
    return preprocessed_corpus
```

3. Construire l’index inversé : on se contente du numéro du document :

```python
from typing import List, Dict, Set

def build_inverted_index(corpus_tokens: List[List[str]]) -> Dict[str, List[int]]:
    """  Builds a simple inverted index.It returns:  Dictionary mapping token -> sorted document IDs """
    inverted_index = {}

    for doc_id, tokens in enumerate(corpus_tokens):
        for token in tokens:
            if token not in inverted_index:
                inverted_index[token] = set()
            inverted_index[token].add(doc_id)
    
    # Convert sets to sorted lists
    for token in inverted_index:
        inverted_index[token] = sorted(inverted_index[token])

    return inverted_index
```

**VarByte Compression :**

```python
from typing import List

def vbe_compress(numbers: List[int]) -> bytes:
    """Compress a list of integers using classic VBE (MSB=stop bit)."""
    result = bytearray()
    for num in numbers:
        bytes_list = []
        while True:
            b = num & 0b01111111  # take 7 least significant bits
            bytes_list.insert(0, b)  # prepend
            num >>= 7
            if num == 0:
                break
        # Set stop bit (MSB) on the last byte
        bytes_list[-1] |= 0b10000000
        result.extend(bytes_list)
    return bytes(result)

def vbe_decompress(data: bytes) -> List[int]:
    """Decompress VBE bytes into a list of integers."""
    numbers = []
    current = 0
    for byte in data:
        if byte & 0b10000000:  # stop bit = 1 → last byte
            current = (current << 7) | (byte & 0b01111111)
            numbers.append(current)
            current = 0
        else:  # stop bit = 0 → more bytes follow
            current = (current << 7) | byte
    return numbers
```

**Gap Encoding :**

Gamma(n):
1. Let b = number of bits to represent n
2. Prefix = (b-1) 1s + 0 (stop bit)
3. Offset = remaining (b-1) bits of n
4. Code = prefix + offset


```python
def delta_encode(sorted_ids):
    """Apply delta encoding to a sorted posting list."""
    if not sorted_ids:
        return []
    return [sorted_ids[0]] + [sorted_ids[i] - sorted_ids[i-1] for i in range(1, len(sorted_ids))]

def delta_decode(deltas):
    """Decode delta-encoded list back to absolute document IDs."""
    if not deltas:
        return []
    output = [deltas[0]]
    for i in range(1, len(deltas)):
        output.append(output[-1] + deltas[i])
    return output
```

**Gamma encoding:**

the problem is how to encode 0 , and 1 ??  

we choose to encode :
- 0 => 0 
- 1 => 10 
- 2 => 100 following the rule   

Still the problem persist :  the 2 is always decoded as 1, 0  

We choose then to avoid any conflicts Shift all numbers by 2  

so, we will : 
- add 2 to the number 
- calculate the b : the necessary number of bits to encode the number 
- create the prefix 1*(b-1) + 0
- add the offset : the last b-1 dits of the number  

```python
from typing import List

def gamma_encode_number(n: int) -> str:
    """Gamma encode n (n >= 2) using stop-bit convention."""
    n += 2  # shift to avoid 0/1 conflicts
    binary = bin(n)[2:]
    b = len(binary)
    prefix = '1' * (b - 1) + '0'
    offset = binary[1:]
    return prefix + offset

def gamma_encode_list(numbers: List[int]) -> int:
    """Encode list of integers"""
    return int(''.join(gamma_encode_number(n) for n in numbers),2)   

def gamma_decode_bits(payload: int) -> List[int]:
    """Decode gamma-encoded bits, then shift -2 to recover original numbers."""
    bits = str(bin(payload)[2:])  # convert to binary string

    numbers = []
    i = 0
    while i < len(bits):
        length = 0
        while i < len(bits) and bits[i] == '1':
            length += 1
            i += 1
        i += 1  # skip stop-bit 0

        offset_bits = bits[i:i+length]
        i += length

        number_bin = '1' + offset_bits
        number = int(number_bin, 2)
        numbers.append(number - 2)  # shift back
    return numbers
```

**VBE + Delta :** 

```python
def vbe_delta_compress(index: Dict[str, List[int]]) -> Dict[str, bytes]:
    """Compress an inverted index using Delta Encoding + VarByte."""
    compressed_index = {}
    for term, doc_ids in index.items():
        gap_compress_index = delta_encode(doc_ids)
        vbe_compress_index = vbe_compress(gap_compress_index)
        compressed_index[term] = vbe_compress_index

    return compressed_index
```

**Gamma + Delta :**

```python
def gamma_delta_compress(index: Dict[str, List[int]]) -> Dict[str, bytes]:
    """Compress an inverted index using Delta Encoding + VarByte."""
    compressed_index = {}
    for term, doc_ids in index.items():
        gap_compress_index = delta_encode(doc_ids)
        gamma_compress_index = gamma_encode_list(gap_compress_index)
        compressed_index[term] = gamma_compress_index

    return compressed_index
```

## Index avec opérations de maintenance :

**Add a document :**

```python
import time
from typing import List, Dict, Set

def add_document(compress_index: Dict[str, int], new_doc: str, corpus: List[str] = corpus) -> Dict[str, int]:
    """Adds a document to the inverted index."""
    doc_id = len(corpus)  # Assuming doc_id is the next document ID added to the corpus 
    corpus.append(new_doc)
    doc_tokens = preprocess_document(new_doc)
    for token in doc_tokens:
        if token not in compress_index:
            compress_index[token] = gamma_encode_list([doc_id])
        else: 
            doc_ids_index = gamma_decode_bits(compress_index[token])
            doc_ids_index.append(doc_id)
            compress_index[token] = gamma_encode_list(doc_ids_index)
    return compress_index
```

**Delete a document (lazy deletion) :**

```python
deleted_docs_ids = set()
```

```python
def delete_document_id(index: Dict[str, int], doc_id: int, deleted_docs_ids: Set[int] = deleted_docs_ids) -> Dict[str, int]:
    deleted_docs_ids.add(doc_id)
    return index

def delete_document(index: Dict[str, int], doc: str, deleted_docs_ids: Set[int] = deleted_docs_ids) -> Dict[str, int]:
    """Marks a document as deleted (lazy deletion)."""
    doc_id = corpus.index(doc)
    return  delete_document_id(index,doc_id,deleted_docs_ids)

def clean_index(index: Dict[str, int], deleted_docs_ids: Set[int]) -> Dict[str, int]:
    """Cleans the inverted index by removing deleted document IDs."""
    cleaned_index = {}
    for token, encoded_postings in index.items():
        doc_ids = gamma_decode_bits(encoded_postings)
        # Remove deleted document IDs
        cleaned_doc_ids = [doc_id for doc_id in doc_ids if doc_id not in deleted_docs_ids]
        if cleaned_doc_ids:
            cleaned_index[token] = gamma_encode_list(cleaned_doc_ids)
    deleted_docs_ids.clear()  # Clear the set after cleaning
    return cleaned_index
```

```python
start_time = time.perf_counter()
delete_document(compress_index, new_doc)
new_index = clean_index(compress_index, deleted_docs_ids)
end_time = time.perf_counter()

initial_index_creation_time = end_time - start_time
print("add document time (seconds):", initial_index_creation_time)
```

**Update a document :**

```python
def update_document_id(index: Dict[str, int], doc_id: int, updated_doc: str, deleted_docs_ids: Set[int] = deleted_docs_ids, corpus: List[str] = corpus) -> Dict[str, int]:
    """Updates a document by deleting the old one and adding the new one."""
    delete_document_id(index, doc_id, deleted_docs_ids)
    new_index = add_document(index, updated_doc,corpus)
    return new_index

def update_document(index: Dict[str, int], doc: str, updated_doc: str, deleted_docs_ids: Set[int] = deleted_docs_ids, corpus: List[str] = corpus) -> Dict[str, int]:
    """Updates a document by deleting the old one and adding the new one."""
    doc_id = corpus.index(doc)
    return update_document_id(index, doc_id, updated_doc, deleted_docs_ids, corpus)
```

## Index avec construction parallélisée :

**Term-oriented / horizontal partitioning :**

Idea: Split the index by terms.

Each machine/node is responsible for a subset of the dictionary.

```python
from typing import Dict, Tuple

def split_dict_into_n(d: Dict[str, int], n: int = 3) -> Tuple[Dict[str,int], ...]:
    """ Splits a dictionary into n smaller dictionaries   """
    items = list(d.items())
    size = len(items) // n
    remainder = len(items) % n

    partitions = []
    start = 0
    for i in range(n):
        end = start + size + (1 if i < remainder else 0)  # distribute remainder
        partitions.append(dict(items[start:end]))
        start = end
    return tuple(partitions)

# Example usage
d1, d2, d3 = split_dict_into_n(compress_index, n=3)

print("Initial dict size:",len(compress_index),"Partition sizes:", len(d1), len(d2), len(d3))
```
