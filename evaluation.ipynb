# define our models  : 

```python
from corpus import Corpus
from model_bm25 import BM25Model
from model_boolean_extended import ExtendedBooleanModel
from model_language import LanguageModel
from vectorial import SmartNotation


bool_model_ntmax = ExtendedBooleanModel(Corpus(), p=2.0, corpusSmart=SmartNotation.N().T().Max())
bool_model_bnn = ExtendedBooleanModel(Corpus(), p=2.0, corpusSmart=SmartNotation.B().N().N())
bool_model_bnc = ExtendedBooleanModel(Corpus(), p=2.0, corpusSmart=SmartNotation.B().N().C())
bool_model_ntmax_4 = ExtendedBooleanModel(Corpus(), p=4.0, corpusSmart=SmartNotation.N().T().Max())

model_25Model = BM25Model(Corpus(), k1=1.5, b=0.75)
lm_model = LanguageModel(Corpus())
```

## define our evaluator: 

```python
from evaluation import Evaluation


evaluator = Evaluation(Corpus())
queries_llm_judged = evaluator.get_queries_with_scores(5)
for query, scores in queries_llm_judged.items():
        print(f"Query: {query}\nScores: {scores}\n")

important_docs = evaluator.get_queries_with_important_docs(5)
for query, docs in important_docs.items():
        print(f"Query: {query}\nImportant Docs: {docs}\n")
```
